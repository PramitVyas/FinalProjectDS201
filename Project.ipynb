```python
import pandas as pd
import numpy as np
stats21 = pd.read_csv('2021stats.csv')
two21 = pd.read_csv('2021allstars.csv')
def newyeardf(df, allstars):
    x = []
    y = []
    for i in df['Player']:
        x.append(i)
    for i in x:
        y.append(i.split("\\", 1)[0])
    df['Player'] = y    
    df.assign(Group='A')
    a = []
    def makeitalist(file):
        l1st = []
        for i in file['Players']:
            l1st.append(i)
        return l1st
    allstars2021 = makeitalist(allstars)
    for i in df['Player']:
        for j in allstars2021:
            if i == j:
                a.append(True)
                break
            else:
                continue
        condition = i in allstars2021
        if condition != True:
            a.append(False)
    df['Allstar'] = a
    
    return df

_2021stats_ = newyeardf(stats21, two21)
_2021stats_['Season'] = '2020-2021'
_2021stats_.sort_values("PTS", ascending = False)
```
