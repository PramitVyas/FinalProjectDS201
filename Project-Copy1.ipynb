```python
import pandas as pd
import numpy as np
stats21 = pd.read_csv('2021stats.csv')
two21 = pd.read_csv('2021allstars.csv')
per100_2021 = pd.read_csv('2021per.csv')[['Player','ORtg','DRtg']].reset_index()
adv2021 = pd.read_csv('2021advances.csv')
```

```python
def newyeardf(df, allstars,rtg,adv):
    x = []
    y = []
    for i in df['Player']:
        x.append(i)
    for i in x:
        y.append(i.split("\\", 1)[0])
    for i in y:
        index = y.index(i)
        if y[index] == y[index-1]:
            df = df.drop([df.index[i]])
            y = y.drop([y.index[i]])
            
    df['Player'] = y    
    df.assign(Group='A')
    a = []
    def makeitalist(file):
        l1st = []
        for i in file['Players']:
            l1st.append(i)
        return l1st
    allstars2021 = makeitalist(allstars)
    for i in df['Player']:
        for j in allstars2021:
            if i == j:
                a.append(True)
                break
            else:
                continue
        condition = i in allstars2021
        if condition != True:
            a.append(False)
    df['Allstar'] = a
    c = []
    d = []
    for i in rtg['Player']:
        c.append(i)
    for i in c:
        d.append(i.split("\\", 1)[0])
    rtg['Player'] = y    
    rtg.assign(Group='A')
    #df = pd.merge(df,rtg)
    
    df.sort_values("Player", ascending = True).reset_index()
    return df

_2021stats_= newyeardf(stats21, two21,per100_2021,adv2021)
_2021stats_['Season'] = '2020-2021'
_2021stats_.head(30)
```
