```python
def newyeardf(df, allstars,rtg,adv):
    df['Player'] = df['Player'].map(lambda x: x.rstrip('*'))
    rtg = rtg[['Player','G','ORtg','DRtg']].reset_index()
    adv = adv[['Player', 'G','PER', 'TS%', '3PAr',
       'FTr', 'ORB%', 'DRB%', 'TRB%', 'AST%', 'STL%', 'BLK%', 'TOV%', 'USG%',
       'OWS', 'DWS', 'WS', 'WS/48', 'OBPM', 'DBPM', 'BPM', 'VORP']]
    x = []
    y = []
    for i in df['Player']:
        x.append(i)
    for i in x:
        y.append(i.split("\\", 1)[0])
    def dropduplicates(fd):
        fd.sort_values("G", ascending = True).reset_index()
        fd = fd.drop_duplicates('Player', keep='first')    
        return fd
    df['Player'] = y    
    df.assign(Group='A')
    a = []
    def makeitalist(file):
        l1st = []
        for i in file['Players']:
            l1st.append(i)
        return l1st
    allstars2021 = makeitalist(allstars)
    for i in df['Player']:
        for j in allstars2021:
            if i == j:
                a.append(1)
                break
            else:
                continue
        condition = i in allstars2021
        if condition != True:
            a.append(0)
    df['Allstar'] = a
    c = []
    d = []
    df = dropduplicates(df)
    for i in rtg['Player']:
        c.append(i)
    for i in c:
        d.append(i.split("\\", 1)[0])
    rtg['Player'] = d  
    rtg.assign(Group='A')
    rtg = dropduplicates(rtg)
    adv = dropduplicates(adv)
    df = pd.merge(df,rtg)
    adv = adv.drop('G', 1)
    adv2 = adv.drop('Player', 1)
    df.reset_index()
    hi = adv2.reset_index()
    df = df.join(hi, lsuffix='_left')
    df = df.drop('index_left', 1)
    df = df.drop('index', 1)
    return df

_2021stats_= newyeardf(stats21, two21,per100_2021,adv2021)
_2021stats_['Season'] = '2020-2021'
_2021stats_.head(30)
```
